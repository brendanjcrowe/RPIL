# Generating Results for Final Project

```python
# Handling data and computation
import numpy as np

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Statistical tools
import scipy.stats

# Modeling
from mixture_of_logistic_models import AdversarialLogits, Adversary_, Expert_
from sklearn.linear_model import LogisticRegression

# Generating data and interacting with environments
from experts import Expert
import gym

# Supressing warnings
# This is bad practices, I am only doing this to keep the notebook clean
import warnings
warnings.filterwarnings('ignore')
```

```python
car = gym.make('MountainCar-v0')
lander = gym.make('LunarLander-v2')
acrobot = gym.make('Acrobot-v1')

car_expert = Expert('MountainCar-v0')
lander_expert = Expert('LunarLander-v2')
acrobot_expert = Expert('Acrobot-v1')
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h
```

## Mountian Car

```python
rewards = []

for i in range(20):
    
    observation = car.reset()
    done = False
    rd = 0
    
    while not done:
        
        observation, reward, done, _ = car.step(car_expert.policy(observation))
        rd += reward
        
    rewards.append(rd)
    
mean_confidence_interval(rewards)
```

```python
perfect_data, _, _ = car_expert.generate_data(20)
fig, ax = plt.subplots(figsize=(10,7))
ax = sns.scatterplot(x=perfect_data[:,0], y=perfect_data[:,1], hue=perfect_data[:,2], palette=['red', 'blue'])
ax.set_title('Expert Data')
ax.set_xlabel('Position')
ax.set_ylabel('Velocity')
```

### Corrupted Data

```python
data1, _, splits1 = car_expert.generate_data(10)
data2, _, splits2 = car_expert.generate_data(10, 'corrupt')
```

```python
start = 0
data = np.vstack([data1, data2])
demos = []
for split in splits1:
    demos.append([start, split-1])
    start = split

splits2 += splits1[-1]
for split in splits2:
    demos.append([start, split-1])
    start = split
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax = sns.scatterplot(x=data[:,0], y=data[:,1], hue=data[:,2], palette=['red', 'green', 'blue'])
ax.set_title('Corrupted Data')
ax.set_xlabel('Position')
ax.set_ylabel('Velocity')
```

```python
model = LogisticRegression(penalty='none').fit(data[:,0:2], data[:,2])
fig, ax = plt.subplots(figsize=(10,7))
ax = sns.scatterplot(x=data[:,0], y=data[:,1], hue=model.predict(data[:,0:2]), palette=['red', 'blue'])
ax.set_title('Learned Policy Assuming Correct Data')
ax.set_xlabel('Position')
ax.set_ylabel('Velocity')
```

```python
rewards = []

for i in range(20):
    
    observation = car.reset().reshape(1, -1)
    done = False
    rd = 0
    
    while not done:
        
        observation, reward, done, _ = car.step(model.predict(observation).astype(int)[0])
        observation = observation.reshape(1, -1)
        rd += reward
        
    rewards.append(rd)
    
mean_confidence_interval(rewards)
```

```python
model = AdversarialLogits( verbose=False)
model.fit(data[:,0:2], data[:,2], demos, 0.2)
expert, adversary = model.result()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax = sns.scatterplot(x=data[:,0], y=data[:,1], hue=expert.model.predict(data[:,0:2]), palette=['red', 'blue'])
ax.set_title('Learned Policy')
ax.set_xlabel('Position')
ax.set_ylabel('Velocity')
```
