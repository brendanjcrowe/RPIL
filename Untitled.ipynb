```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = np.random.multivariate_normal([5, 5], [[0.5, 0],[0, 10]], 100)
b = np.random.multivariate_normal([5, 5], [[10, 0],[0, 0.5]], 100)
c = np.random.multivariate_normal([10, 10], [[3,0.1],[1, 2]], 100)
d = np.concatenate((a, b, c), axis=0)
data = pd.DataFrame(d, columns=['x', 'y'])
data
data['g'] = np.concatenate((np.zeros(100), np.ones(100), np.full(100, 2)), axis=0)
```

```python
data
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
sns.scatterplot('x', 'y', hue='g',data=data, palette=['red', 'blue', 'green'], ax=ax)
```
